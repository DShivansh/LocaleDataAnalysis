```python
import folium
import pandas as pd
import numpy as np
from collections import Counter
from folium import plugins
```

```python
import sys
import os
import traceback
import optparse
import time
import logging


def wavelength_to_rgb(wavelength, gamma=0.8):

    wavelength = float(wavelength)
    if wavelength >= 380 and wavelength <= 440:
        attenuation = 0.3 + 0.7 * (wavelength - 380) / (440 - 380)
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif wavelength >= 440 and wavelength <= 490:
        R = 0.0
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        B = 1.0
    elif wavelength >= 490 and wavelength <= 510:
        R = 0.0
        G = 1.0
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif wavelength >= 510 and wavelength <= 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0
        B = 0.0
    elif wavelength >= 580 and wavelength <= 645:
        R = 1.0
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif wavelength >= 645 and wavelength <= 750:
        attenuation = 0.3 + 0.7 * (750 - wavelength) / (750 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    else:
        R = 0.0
        G = 0.0
        B = 0.0
    R *= 255
    G *= 255
    B *= 255
    
    return '#%02x%02x%02x' % (int(R), int(G), int(B))

```

```python
m = folium.Map(location=[12.9750, 77.5835], zoom_start=12)
```

```python
df = pd.read_csv('data.csv')
```

```python
source_arr = list(zip(df['from_lat'], df['from_long']))
```

```python
freq = Counter(df['from_area_id'])
```

```python
minimum = min(freq.values())
maximum = max(freq.values())
```

```python
for i, j in source_arr:
    if(not np.isnan(i) and not np.isnan(j)):
        area_id = int(df[(df['from_lat'] == i) & (df['from_long']==j)][:1]['from_area_id'])
        freqen = freq[area_id]
        adjusted_freq = 100*(freqen-minimum)/(maximum-minimum) + 490
        folium.Marker([i,j], icon=plugins.BeautifyIcon(text_color='#000', border_color='transparent',background_color=wavelength_to_rgb(adjusted_freq))).add_to(m)
        print(wavelength_to_rgb(adjusted_freq))
m.save("map2.html")
```
